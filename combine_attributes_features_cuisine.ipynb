```python
import pandas as pd
```

```python
features_df = pd.read_pickle('final_new_df_features.pkl')
features_df.head()
```

```python
chain_df = pd.read_pickle("is_chain_feature.pkl")
chain_df.head()
```

```python
def check_is_chain(name):
    size = len(features_df[features_df['name'] == name])
    return size > 1

features_df['is_chain'] = features_df['name'].apply(check_is_chain)
features_df.head()
```

```python
attributes_df = pd.read_pickle("df_with_attributes.pkl")
attributes_df.head()
```

```python
cuisine_df = pd.read_csv("cusine_info_updated.csv")
cuisine_df = cuisine_df[['id', 'cuisine']]
cuisine_df.head()
```

```python
merge_attr_features_df = pd.merge(features_df, attributes_df, on=['id'], how='inner')
merge_attr_features_df.head()
```

```python
merge_attr_features_cuisine_df = pd.merge(merge_attr_features_df, cuisine_df, on=['id'], how='inner')
merge_attr_features_cuisine_df.head()
```
