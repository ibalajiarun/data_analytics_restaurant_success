```python
import pandas as pd
```

```python
YELP_NEW_RESTAURANT_DF_PKL = '../data/new_restaurant.pkl'
```

```python
final_new_df = pd.read_pickle('final_new_df.pkl')
final_new_df.head()
```

```python
new_restaurant_df = pd.read_pickle(YELP_NEW_RESTAURANT_DF_PKL)
new_restaurant_df.rename(columns={"business_id": "id"}, inplace=True)
new_restaurant_df.head()
```

```python
print(len(final_new_df))
print(len(new_restaurant_df))
```

```python
merge_df = pd.merge(new_restaurant_df[['id', 'attributes']], final_new_df, on=['id'], how='inner')
merge_df.head()
```

```python
print(len(merge_df))
```

```python
unmerge_df = pd.merge(new_restaurant_df[['id', 'attributes']], final_new_df, on=['id'], how='outer', indicator=True)
unmerge_df = unmerge_df[unmerge_df['_merge'] == 'right_only'].reset_index()
unmerge_df.head()
```
