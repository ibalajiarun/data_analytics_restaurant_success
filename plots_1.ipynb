```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
final_new_df = pd.read_pickle('final_data_with_features_attributes.pkl')
```

```python
final_new_df.columns
```

```python
closed = final_new_df.loc[final_new_df['is_closed'] == True]
open = final_new_df.loc[final_new_df['is_closed'] == False]
```

```python
plt.hist([closed['rating'], open['rating']], label=['closed', 'open'])
plt.legend()
```

```python
plt.hist([closed['rating'], open['rating']], label=['closed', 'open'], normed=True)
plt.legend()
```

```python
plt.hist([closed['price_ordinal'], open['price_ordinal']], label=['closed', 'open'], normed=True)
plt.legend()
```

```python
plt.hist([closed['income'], open['income']], label=['closed', 'open'], normed=True)
plt.legend()
```

```python
plt.hist([closed['is_fastfood'], open['is_fastfood']], label=['closed', 'open'])
plt.legend()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(open['coordinates.latitude'], open['coordinates.longitude'], s=10, c='r', label='open')
ax1.scatter(closed['coordinates.latitude'], closed['coordinates.longitude'], s=10, c='b', label='closed')
plt.legend(loc='lower left');
ax1.set_ylim([-113,-111])
plt.show()
# plt.scatter(final_new_df['coordinates.latitude'], final_new_df['coordinates.longitude'])
```
