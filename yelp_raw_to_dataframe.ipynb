```python
import pandas as pd
```

```python
YELP_OLD_BUSINESS_SRC = '../data/old_business.json'
YELP_NEW_BUSINESS_SRC = '../data/new_business.json'

YELP_OLD_RESTAURANT_SRC = '../data/old_restaurant.json'
YELP_NEW_RESTAURANT_SRC = '../data/new_restaurant.json'

YELP_OLD_RESTAURANT_DF_PKL = '../data/old_restaurant.pkl'
YELP_NEW_RESTAURANT_DF_PKL = '../data/new_restaurant.pkl'

YELP_MERGED_RESTAURANT_DF_PKL = '../data/merged_restaurant.pkl'

YELP_UNMERGED_CSV = "../data/unmerged_restaurant.csv"
YELP_UNMERGED_BALAJI_CSV = "../data/unmerged_restaurant_balaji.csv"

YELP_FINAL_UNMERGED_PKL = "../data/final_unmerged.pkl"
```

```python
def line_count(file_path):
    with open(file_path) as file:
        for i, l in enumerate(file):
            pass
    return i + 1
```

```python
print ("Number of businesses in old dataset: " , line_count(YELP_OLD_BUSINESS_SRC))
print ("Number of businesses in new dataset: " , line_count(YELP_NEW_BUSINESS_SRC))
```

```python
def extract_restaurants(file_path, new_file_path):
    count = 0
    with open(file_path) as file, open(new_file_path, "w") as out_file:
        for i, l in enumerate(file):
            if "Restaurant" in l and '"state": "AZ"' in l:
                out_file.write(l)
                count = count + 1 
    return count

print ("Number of old businesses: " , extract_restaurants(YELP_OLD_BUSINESS_SRC, YELP_OLD_RESTAURANT_SRC))
print ("Number of new businesses: " , extract_restaurants(YELP_NEW_BUSINESS_SRC, YELP_NEW_RESTAURANT_SRC))
```

```python
print ("Number of restaurant in old dataset: " , line_count(YELP_OLD_RESTAURANT_SRC))
print ("Number of restaurant in new dataset: " , line_count(YELP_NEW_RESTAURANT_SRC))
```

```python
old_restaurant_df = pd.read_json(YELP_OLD_RESTAURANT_SRC, lines=True)

# replaces the new line character in full_address column in old data set with comma
old_restaurant_df['full_address'] = old_restaurant_df['full_address'].str.replace("\n", ", ")

old_restaurant_df = old_restaurant_df.loc[old_restaurant_df['open'] == True]

old_restaurant_df.head()
```

```python
new_restaurant_df = pd.read_json(YELP_NEW_RESTAURANT_SRC, lines=True)

# add a new full_address column for compatibility with old_restaurant_df
new_restaurant_df['full_address'] = new_restaurant_df['address'] + ", " + new_restaurant_df['city'] + ", " + \
                                    new_restaurant_df['state'] + " " + new_restaurant_df['postal_code']

new_restaurant_df.head()
```

```python
# write the df to file
old_restaurant_df.to_pickle(YELP_OLD_RESTAURANT_DF_PKL)
new_restaurant_df.to_pickle(YELP_NEW_RESTAURANT_DF_PKL)

print ("Number of restaurant in old dataset: " , len(old_restaurant_df))
print ("Number of restaurant in new dataset: " , len(new_restaurant_df))
```

```python
old_restaurant_info_df = old_restaurant_df[['business_id', 'name', 'full_address']]
```

```python
merge_df = pd.merge(old_restaurant_info_df, new_restaurant_df, on=['name', 'full_address'], how='inner')
print(len(merge_df.index))
merge_df.head()
```

```python
# write the df to file
merge_df.to_pickle(YELP_MERGED_RESTAURANT_DF_PKL)
```

```python
outer_merge_df = pd.merge(old_restaurant_info_df, new_restaurant_df, on=['name', 'full_address'], how='outer', indicator=True)
unmerged_df = outer_merge_df[outer_merge_df['_merge'] == 'left_only'].reset_index()
print(len(unmerged_df))
unmerged_df.head()
```

```python
df = pd.merge(unmerged_df[['name', 'full_address']], new_restaurant_df, on=['full_address'], how='inner')
# df = new_restaurant_df.loc[new_restaurant_df['full_address'].isin(unmerged_df['full_address'])]
df[['name_x', 'name_y', 'full_address', 'is_open']].head()
```
