```python
import pandas as pd
import numpy as np
```

```python
YELP_OLD_RESTAURANT_SRC = '../data/old_restaurant.json'
YELP_NEW_RESTAURANT_SRC = '../data/new_restaurant.json'

YELP_OLD_BUSINESS_SRC = '../data/old_business.json'
YELP_NEW_BUSINESS_SRC = '../data/new_business.json'
```

```python
def extract_restaurants(file_path, new_file_path):
    count = 0
    with open(file_path) as file, open(new_file_path, "w") as out_file:
        for i, l in enumerate(file):
            if "Restaurant" in l and '"state": "AZ"' in l:
                out_file.write(l)
                count = count + 1 
    return count

print ("Number of old businesses: " , extract_restaurants(YELP_OLD_BUSINESS_SRC, YELP_OLD_RESTAURANT_SRC))
print ("Number of new businesses: " , extract_restaurants(YELP_NEW_BUSINESS_SRC, YELP_NEW_RESTAURANT_SRC))
```

```python
old_restaurant_df = pd.read_json(YELP_OLD_RESTAURANT_SRC, lines=True)
```

```python
old_restaurant_df['full_address'] = old_restaurant_df['full_address'].str.replace("\n", ", ")

old_restaurant_df = old_restaurant_df.loc[old_restaurant_df['open'] == True]

old_restaurant_df.head()
```

```python
cities = old_restaurant_df['city'].unique()
np.sort(cities)
```

```python
new_restaurant_df = pd.read_json(YELP_NEW_RESTAURANT_SRC, lines=True)

new_restaurant_df = new_restaurant_df.loc[new_restaurant_df['categories'].apply(lambda x : True if "Restaurants" in x else False)]

# add a new full_address column for compatibility with old_restaurant_df
new_restaurant_df['full_address'] = new_restaurant_df['address'] + ", " + new_restaurant_df['city'] + ", " + \
                                    new_restaurant_df['state'] + " " + new_restaurant_df['postal_code']

new_restaurant_df.head()
```
