```python
import pandas as pd
import json
```

```python
merged_restaurant_latest_info_df = pd.read_json('merged_restaurant_latest_info_df.json', typ='Series')
merged_restaurant_latest_info_df.head()
```

```python
merged_restaurant_new_df = pd.io.json.json_normalize(merged_restaurant_latest_info_df.apply(json.loads))
merged_restaurant_new_df.head()
```

```python
merged_restaurant_new_df.columns
```

```python
final_merged_new_df = merged_restaurant_new_df[['alias', 'categories', 'coordinates.latitude', 'coordinates.longitude',
       'hours', 'id',
       'is_closed', 
       'location.display_address', 'location.zip_code',
       'name', 'price', 'rating', 'review_count']]
```

```python
final_merged_new_df.to_pickle('final_merged_new_df.pkl')
```

```python
all_unmerged_final_latest_info_df = pd.read_json('all_unmerged_final_latest_info_df.json', typ='Series')
all_unmerged_final_latest_info_df.head()
```

```python
all_unmerged_restaurant_new_df = pd.io.json.json_normalize(all_unmerged_final_latest_info_df.apply(json.loads))
all_unmerged_restaurant_new_df.head()
```

```python
final_unmerged_new_df = all_unmerged_restaurant_new_df[['alias', 'categories', 'coordinates.latitude', 'coordinates.longitude',
       'hours', 'id',
       'is_closed', 
       'location.display_address', 'location.zip_code',
       'name', 'price', 'rating', 'review_count']]
```

```python
final_unmerged_new_df.to_pickle('final_unmerged_new_df.pkl')
```

```python
final_new_df = pd.concat([final_unmerged_new_df, final_merged_new_df])
print(len(final_new_df))
final_new_df.head()
```

```python
final_new_df.dropna(subset=['alias'], inplace=True)
print(len(final_new_df))
```

```python
final_new_df.to_pickle("final_new_df.pkl")
```
