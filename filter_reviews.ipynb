```python
import pandas as pd
```

```python
final_merged_new_df = pd.read_pickle('final_merged_new_df.pkl')
```

```python
def extract(match_list, file_path, new_file_path):
    count = 0
    with open(file_path) as file, open(new_file_path, "w") as out_file:
        for i, l in enumerate(file):
            for b_id in match_list:
                if b_id in l:
                    out_file.write(l)
                    count = count + 1 
    return count
```

```python
match_list = final_merged_new_df['id'].dropna().values.tolist()
len(match_list)
```

```python
extract(match_list, "../data/all_business_reviews_new.json", "../data/new_reviews.json")
```
